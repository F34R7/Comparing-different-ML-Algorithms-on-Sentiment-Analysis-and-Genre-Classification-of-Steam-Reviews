```python
import requests
import pandas
from urllib.parse import quote
```

```python
url = "https://store.steampowered.com/appreviews/"

game_id = "1677280"
appendix = "?json=1"
language = "&language=english"
#filter2 = "&filter=recent"
cursor_appendix = "&cursor="
cursor = "*"
result = []

#1000 is max for 1 day, 5000 for 100.000 reviews
for i in range (0,1000):
    url2 = url + game_id + appendix + language + cursor_appendix + cursor
    response = requests.get(url2)
    if response.status_code == 200:
        reviews_data = response.json()
        
    else:
        print("Error", response.status_code)
    try:
        cursor = reviews_data["cursor"]
        cursor = quote(cursor)
    except: 
        result.append(reviews_data)
        break
    result.append(reviews_data)
```

```python
newlist = []
for i in result:
    for j in i["reviews"]:
        #print(j["review"] + "," + str(j["voted_up"]) +"\n")
        temp_list = [j["review"], j["voted_up"]]
        newlist.append(temp_list)
```

```python
df = pandas.DataFrame(newlist, columns = ["Review","Recommended"])
df["Genre"] = "Strategy"
df
```

```python
df.to_csv("Steam_Reviews", mode = "a", index = False)
```
